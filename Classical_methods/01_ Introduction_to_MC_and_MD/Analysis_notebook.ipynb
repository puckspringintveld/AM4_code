```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import fsolve
TUe_red='#a80707'
```

```python
def plt_potentials(file, sim_type,temp, show_plot):
    data=np.genfromtxt(file)

    conf_number=data[:,0]
    curr_PE=data[:,1]*1.38e-23*6.022e23/1e3 #convert to KJ/mol
    cum_av_PE=data[:,2]*1.38e-23*6.022e23/1e3 #convert to KJ/mol

    if show_plot:
        plt.plot(conf_number,curr_PE, label=f'Current')
        plt.plot(conf_number,cum_av_PE, label=f'Average')
        plt.legend()
        plt.title(f'{sim_type} at {temp}K')
        plt.ylabel('PE (KJ/mol)')
        plt.xlabel('iteration')
        plt.show()
    return conf_number,curr_PE, cum_av_PE
```

```python
it,c_MD_350,av_MD_350=plt_potentials('01_MD_Short_run/U_energy.dat','MD',350, True)
```

```python
it,c_MC_350,av_MC_350=plt_potentials('01_MC_Short_run/U_energy.dat','MC',350, True)
```

```python
plt.plot(it,c_MC_350, label='MC')
plt.plot(it,c_MD_350, label='MD', c=TUe_red)
plt.ylabel('PE (KJ/mol)')
plt.xlabel('Iteration')
plt.title('Comparison of Current Energies for MD and MC')
plt.legend()
plt.savefig('01_curr_MD_vs_MC',dpi=300)
plt.show()
```

```python
plt.plot(it,av_MC_350, label='MC')
plt.plot(it,av_MD_350, label='MD', c=TUe_red)
plt.ylabel('PE (KJ/mol)')
plt.xlabel('Iteration')
plt.title('Comparison of Average Energies for MD and MC')
plt.legend()
plt.savefig('01_av_MD_vs_MC')
plt.show()
```
