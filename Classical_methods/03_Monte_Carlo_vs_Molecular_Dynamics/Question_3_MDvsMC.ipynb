```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data_MD=np.genfromtxt('03_MD_RDFs/RadialDistributionFunctions/System_0/RDF_CH4_amm_CH4_amm.dat')
data_MC=np.genfromtxt('03_MC_RDFs/RadialDistributionFunctions/System_0/RDF_CH4_amm_CH4_amm.dat')
```

```python
index=data_MD[:,0]
distance_MD=data_MD[:,1]
RDF_MD=data_MD[:,2]
n_RDF_MD=data_MD[:,3]

index=data_MC[:,0]
distance_MC=data_MC[:,1]
RDF_MC=data_MC[:,2]
n_RDF_MD=data_MC[:,3]
```

```python
plt.plot(distance_MD, RDF_MD, label='MD')
plt.plot(distance_MC,RDF_MC, label='MC')
plt.legend()
```

```python
ref_data_MD=np.genfromtxt('03_MD_RDFs/reference_RDF_CH4_amm_CH4_amm.dat')
```

```python
index=ref_data_MD[:,0]
distance_ref=ref_data_MD[:,1]
RDF_ref=ref_data_MD[:,2]
n_RDF_ref=ref_data_MD[:,3]
```

```python
plt.plot(distance_ref,RDF_ref,label='Reference', c='k')
plt.plot(distance_MD, RDF_MD, label='MD', alpha=0.7)
plt.plot(distance_MC,RDF_MC ,label='MC', alpha=0.7)
plt.xlabel('Distance ()')
plt.ylabel('RDF')
plt.legend()
plt.savefig('03_RDF_comp',dpi=300)
plt.show()
```

```python
RMSE_MD=np.sqrt(np.sum((RDF_MD-RDF_ref)**2))
RMSE_MC=np.sqrt(np.sum((RDF_MC-RDF_ref)**2))
print(f'RMSE for MD = {RMSE_MD}, RMSE for MC = {RMSE_MC}')
```
