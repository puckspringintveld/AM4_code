```python
import numpy as np
import matplotlib.pyplot as plt
TUe_red='#a80707'
```

```python
data=np.genfromtxt('360/Density.dat')
```

```python
N_cycles=data[:,0]
Density=data[:,1]
Dens_cum_av=data[:,2]
```

```python
plt.plot(N_cycles ,5.44799*np.ones(len(N_cycles)),'--',label='mean value' ,c='k')
plt.plot(N_cycles,Density,':' ,label='current desnity', alpha=0.6)
plt.plot(N_cycles, Dens_cum_av, label='cumulative average', c=TUe_red)
plt.legend()
plt.xlabel('MC Cycles')
plt.ylabel(r'Desnity ($\frac{kg}{m^3}$)')
plt.savefig('04_density',dpi=300)
plt.show()
```

```python
print('the average found in simulation output file for 360K is:  5.44799 [kg/m^3] +/-0.05996 [kg/m^3]')
```

```python
T=np.array([300,320,340,360,400,500,600]) #K
Av_rho_calc=np.array([6.63951,6.20394,5.76222,5.44799, 4.91114,3.91327, 3.24365]) #kg/m^3
error_calc=np.array([ 0.07564, 0.07273, 0.09002,0.05996,0.03166,0.00899,0.02281]) #kg/m^3
Av_rho_real=np.array([6.541,6.11, 5.733,5.402,4.845,3.859,3.21])
print(Av_rho_calc-Av_rho_real)
```

```python
mask=np.where(T>350)
plt.plot(T[mask],Av_rho_real[mask], '--*',label='Experiment')
plt.errorbar(T[mask],Av_rho_calc[mask],yerr=error_calc[mask],capsize=3, fmt="--.", ecolor = "black", label='Simulation')
plt.legend()
plt.xlabel('Temperature (K)')
plt.ylabel(r'Desnity ($\frac{kg}{m^3}$)')
plt.savefig('04_desnity_v_T',dpi=300)
plt.show()
```

```python
plt.plot(T,Av_rho_calc-Av_rho_real, 'o')
plt.plot(T,np.zeros_like(T), '--', c='k')
plt.xlabel('Temperature (K)')
plt.ylabel(r'Residuals ($\frac{kg}{m^3}$)')
plt.savefig('04_residuals_300_600',dpi=300)
plt.show()
```

```python
plt.plot(T,Av_rho_real, '--*',label='Experiment')
plt.errorbar(T,Av_rho_calc,yerr=error_calc,capsize=3, fmt="--.", ecolor = "black", label='Simulation')
plt.legend()
plt.xlabel('Temperature (K)')
plt.ylabel(r'Desnity ($\frac{kg}{m^3}$)')
plt.savefig('04_desnity_v_T_300',dpi=300)
plt.show()
```
